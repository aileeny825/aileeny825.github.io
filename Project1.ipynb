### Mean, median, mode using Python

```python
import csv

values = []

with open('/Users/aileen/Downloads/SIPA CS Proj 1/student_habits_performance.csv', 'r', newline='') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        values.append(float(row['social_media_hours']))
```

Mean

```python
def cal_mean(values):
    n = len(values)
    val_sum = sum(values)
    mean = val_sum / n
    return mean

cal_mean(values)
```

Median

```python
def cal_median(values):
    s = sorted(values)
    n = len(s)
    
    # Compute the middle number
    mid = n // 2
    
    # If there is odd number of values, return exactly the middle value
    if n % 2 == 1:
        return s[mid]
    
    # If there is even number of values, return the average of the two middle values
    else:
        return (s[mid - 1] + s[mid]) / 2
    
cal_median(values)
```

Mode

```python
from collections import Counter

def cal_mode(values):
    counts = Counter(values)
    
    max_count = max(counts.values())
    modes = []
    for item, value in counts.items():
        if value == max_count:
            modes.append(item)
    return modes

cal_mode(values)
```

### Data Visualization

```python
def visual_blocks(data, title):
    chars = "▁▂▃▄▅▆▇█"
    
    #Turn data into dictionary
    data = data.to_dict()
    vals = list(data.values())
    
    # Find the smallest and largest value in the dataset
    min_val = min(vals)
    max_val = max(vals)

    print(title)
    
    for label, i in data.items():
        # If the max and min value are the same, the level should be 0
        if max_val == min_val:
            level = 0
        # Scale the level between level 1 and 8
        else:
            level = int((i - min_val) / (max_val - min_val) * (len(chars) - 1))
        
        print(f"{label}: {chars[level]}  {i:.2f} ")
```

### Anlyzing whether participating in extracurriculars will affect exam performance or not

```python
avg_extra = df.groupby('extracurricular_participation')['exam_score'].mean()
visual_blocks(avg_extra, title = "Average exam score by extracurricular participation")
```

Since that the average scores do not differ from each other signifiantly, participation in extracarricular is unlikely to affect exam performance of students.

### Anlyzing whether the quality of diet will affect exam performance or not

```python
avg_by_diet = df.groupby('diet_quality')['exam_score'].mean().sort_values()
visual_blocks(avg_by_diet, title="Average exam score by diet quality")
```

We can see there is a slight positive correlation between the quality of diet and exam scores achieved by students. Students with better diets are likely to perform better. However, the relationship is not very significant as the average exam scores by diet quality do not differ significantly.

### Anlyzing whether the number of hours of study each day will affect exam performance or not

```python
# Group by length of study hours and compute mean exam_score
avg_by_study = (
    df.groupby('study_hours_per_day')['exam_score']
      .mean()
      .sort_index()
)

visual_blocks(avg_by_study, title="Average exam score by study hour (ascending)")
```

As we observe the blocks, we can see for higher study hours, students usually get higher exam scores. We can conclude that students who spend more hours of study per day will likely perform better academically.
